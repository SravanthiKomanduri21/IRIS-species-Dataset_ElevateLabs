```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/IRISDataset.csv')

# Display first rows
df.head()
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

df[numerical_cols].hist(figsize=(10,6), bins=20)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Species', data=df)
plt.title('Count of Each Iris Species')
plt.show()
```

```python
plt.figure(figsize=(10,6))
for i, col in enumerate(numerical_cols):
    plt.subplot(2,2,i+1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Encode target
df['Species_encoded'] = df['Species'].astype('category').cat.codes

corr_with_target = df[numerical_cols].corrwith(df['Species_encoded'])
corr_with_target.sort_values(ascending=False)

```
